# Tennis women data visualization

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# better visuals
sns.set(color_codes=True)  

# You need this for plots to show in notebooks
%matplotlib inline
```

```python
women = pd.read_csv('https://s3-eu-west-1.amazonaws.com/nibble-datasets/tennis/womens_train_file.csv')
women.head(10)
```


## Informations tout type de coup confondu


### Positionnement du gagnant par rapport au filet & la ligne centrale lors du dernier coup, tout type de coup confondu


```python
plt.figure(figsize=(16, 14))
sns.scatterplot(x="player.impact.depth", y="player.impact.distance.from.center", hue="outcome", data=women);
```

## Informations coups gagnants

### Vitesse du dernier coup gagnant

```python
winning = women['outcome'] == 'W'
winning_df = women.loc[winning]

plt.figure(figsize=(16, 3))
winning_speed_boxplot = sns.boxplot(x=winning_df['speed'])
```



### Positionnement du gagnant par rapport Ã  la ligne centrale et au filet pour un coup gagnant

```python
winning_pos_df = pd.DataFrame()
winning_pos_df['player.impact.distance.from.center'] =  winning_df['player.impact.distance.from.center']
winning_pos_df['player.impact.depth'] = winning_df['player.impact.depth']

plt.figure(figsize=(16, 14))
sns.scatterplot(x="player.impact.depth", y="player.impact.distance.from.center", data=winning_pos_df);
```



### Positionnement du gagnant par rapport Ã  la ligne centrale et au filet en fonction du nombre de coups Ã©changÃ©s pour un coup gagnant

```python
winning_pos_df['rally'] = winning_df['rally']

plt.figure(figsize=(16, 14))
sns.scatterplot(x="player.impact.depth", y="player.impact.distance.from.center", hue='rally', data=winning_pos_df);
```



### Positionnement de l'adversaire Ã  la rÃ©ception de la balle par rapport Ã  la ligne centrale et au filet pour un coup gagnant


```python
losing_df = pd.DataFrame()
losing_df['opponent.distance.from.center'] =  winning_df['opponent.distance.from.center']
losing_df['opponent.depth'] = winning_df['opponent.depth']

plt.figure(figsize=(16, 14))
sns.scatterplot(x="opponent.depth", y="opponent.distance.from.center", data=losing_df);
```



### Positionnement de l'adversaire Ã  la rÃ©ception de la balle par rapport Ã  la ligne centrale et au filet pour un coup gagnant selon le nombre d'Ã©changes
